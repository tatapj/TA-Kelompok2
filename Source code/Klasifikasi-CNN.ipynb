```python
def classify(path):
  #predicting images
  #path = fn
  img = image.load_img(path, target_size = (150, 150))
  implot = plt.imshow(img)
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis = 0)
    
  images = np.argmax([x])
  classes = model.predict(x)
  classes = np.argmax(classes)
  if classes == 0:
        print("Kelas: Cerah")
  elif classes == 1:
        print("Kelas: Hujan")
  elif classes == 2:
        print("Kelas: Mendung")

  return classes
```

```python
classify("C:/Users/User/Documents/Tugas-Akhir-Kel02/Result_Dataset/test/hujan/19.jpg")
```

```python
cerah_t = 0
cerah_f = 0
hujan_t = 0
hujan_f = 0
mendung_t = 0
mendung_f = 0
```

```python
for i, ret in enumerate(os.walk('C:/Users/User/Documents/Tugas-Akhir-Kel02/Result_Dataset/test/cerah')):
    for i, filename in enumerate(ret[2]):
        if filename.startswith("."):
            continue
        print("Kelas: Cerah")
        classes = classify(ret[0] + '/' + filename)
        if classes == 0:
            cerah_t += 1
        else:
            cerah_f += 1

for i, ret in enumerate(os.walk('C:/Users/User/Documents/Tugas-Akhir-Kel02/Result_Dataset/test/hujan')):
    for i, filename in enumerate(ret[2]):
        if filename.startswith("."):
            continue
        print("Kelas: Hujan")
        classes = classify(ret[0] + '/' + filename)
        if classes == 1:
            hujan_t += 1
        else:
            hujan_f += 1

for i, ret in enumerate(os.walk('C:/Users/User/Documents/Tugas-Akhir-Kel02/Result_Dataset/test/mendung')):
    for i, filename in enumerate(ret[2]):
        if filename.startswith("."):
            continue
        print("Kelas: Mendung")
        classes = classify(ret[0] + '/' + filename)
        if classes == 2:
            mendung_t += 1
        else:
            mendung_f += 1
            
```

```python
#Check metrics

print("True cerah: ", cerah_t)
print("False cerah: ", cerah_f)
print("True Hujan: ", hujan_t)
print("False Hujan: ", hujan_f)
print("True Mendung: ", mendung_t)
print("False Mendung: ", mendung_f)
```
